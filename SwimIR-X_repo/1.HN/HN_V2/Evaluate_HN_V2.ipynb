```python
import numpy as np
from datasets import load_dataset

# Load your .npy file with hard negative information
hard_negatives_data = np.load('/content/swim_ir_yo_training_batches.npy', allow_pickle=True)

# Dummy dataset for demonstration (replace with your actual dataset)
# This mimics the structure your preprocess_dataset function expects

dataset_name = "nthakur/swim-ir-monolingual"
dataset = load_dataset(dataset_name,'yo', trust_remote_code=True)


# Preprocessing function
def preprocess_dataset(dataset):
    """
    Process the MIRACL dev split to create queries, positives, and corpus.

    Returns:
        queries (dict): {query_id: query_text}
        positives (dict): {query_id: positive_docid}
        corpus (dict): {passage_id: passage_text}
    """
    data = dataset["train"]

    # Create queries dictionary: {query_id: query_text}
    queries = {item["_id"]: item["query"] for item in data}

    # Create positives dictionary: {query_id: positive_docid}
    positives = {item["_id"]: item["_id"] for item in data}

    # Create corpus dictionary: {passage_id: passage_text}
    corpus = {item["_id"]: item["title"]+item["text"] for item in data}

    return queries, positives, corpus

# Preprocess the dataset
queries, positives, corpus = preprocess_dataset(dataset)

# Function to print negative examples
def print_negative_example(hard_negatives_data, index):
    query_id = hard_negatives_data[index]['query_id']
    query_text = hard_negatives_data[index]['query_text']
    hard_negative_ids = hard_negatives_data[index]['hard_negative_ids']

    print(f"\nQuery {index} (ID: {query_id}):")
    print(f"Query text: {query_text}")
    print("Documents:")

    for neg_id in hard_negative_ids:
        if neg_id in corpus:
            print(f"- Document ID: {neg_id}")
            print(f"  Content: {corpus[neg_id]}")
        else:
            print(f"- Document ID: {neg_id} (Not found in corpus)")

# Print the first negative example
print_negative_example(hard_negatives_data, 0)
```
